```python
business = pd.read_csv("yelp_dataset/business.json.csv")
business
```

```python
business.describe()
```

```python
us_states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", 
          "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA",
          "RI", "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]

us_business = business.loc[business['state'].isin(us_states)]

us_business
```

```python
us_business.describe()
```

```python
us_business = us_business[us_business['categories'].isna()!=True]
us_business.describe()
```

```python
#only US restaurants

us_business_restaurant = us_business[us_business['categories'].str.contains('Restaurants')]
us_business_restaurant
```

```python
us_business_restaurant.isnull().sum()
```
