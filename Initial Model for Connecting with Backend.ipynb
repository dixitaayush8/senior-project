```python
import pandas as pd
import numpy as np
from surprise import KNNBasic, evaluate
from surprise import Dataset, Reader
from surprise import SVDpp, model_selection
from sklearn.externals import joblib
from surprise import SVD, NMF
```

```python
us_restaurant_review_lasvegas_nv = pd.read_csv("yelp_dataset/us_restaurant_review_lasvegas_nv.csv")
```

```python
us_restaurant_review_lasvegas_nv
```

# Cosine similarity

```python
reader = Reader(rating_scale = (1, 5))
us_restaurant_review_lasvegas_nv_data = Dataset.load_from_df(us_restaurant_review_lasvegas_nv[['user_id', 'business_id', 'review_stars']], reader)
lasvegas_nv_data_train = us_restaurant_review_lasvegas_nv_data.build_full_trainset()
sim_options = {
    'name': 'cosine',
    'user_based': False
}

knn = KNNBasic(sim_options = sim_options)

knn.fit(lasvegas_nv_data_train)
```

```python
lasvegas_nv_data_test = lasvegas_nv_data_train.build_anti_testset()
predictions = knn.test(lasvegas_nv_data_test)
```

```python
predictions
```

# Matrix Factorization

```python
svdpp = SVDpp()
svdpp.fit(lasvegas_nv_data_train)
```

```python
predictions_svdpp = svdpp.predict('3nDUQBjKyVor5wV0reJChg', '6fPQJq4f_yiq1NHn0fd11Q')
predictions_svdpp
```

### Grab business_ids that user_id 3nDUQBjKyVor5wV0reJChg did not rate in order to recommend restaurants they have not visited


```python
business_ids = us_restaurant_review_lasvegas_nv['business_id'].unique()
business_ids_rated_by_user = us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv['user_id'] == '3nDUQBjKyVor5wV0reJChg', 'business_id']
business_ids_to_predict = np.setdiff1d(business_ids, business_ids_rated_by_user)
```

### Predict the review_ratings for each business_id that user_id 3nDUQBjKyVor5wV0reJChg did not rate

```python
testset = [['3nDUQBjKyVor5wV0reJChg', business_id, 4.] for business_id in business_ids_to_predict]
predictions_testset = svdpp.test(testset)
predictions_testset
```

### Output the top 30 business_ids with the top 30 best predicted review_rating

```python
predicted_ratings = np.array([pred.est for pred in predictions_testset])
index_max_pred_rating = np.argpartition(predicted_ratings,-30)[-30:]
business_id_recommended = business_ids_to_predict[index_max_pred_rating]
print('Top 10 businesses:\n {0}, \npredicted review_ratings: \n{1}'.format(business_id_recommended, predicted_ratings[index_max_pred_rating]))
```

# Untuned Models

## Mean RMSE values
### SVDpp: 1.0095
### SVD: 1.0065
### NMF: 1.0545

## Since SVDpp and SVD have lower RMSE values than NMF, we would tune only SVDpp and SVD as they have greater potential to produce even lower RMSE values, indicating a more accurate model and hence predictions.

```python
svdpp = SVDpp()
model_selection.cross_validate(svdpp, us_restaurant_review_lasvegas_nv_data, verbose = True)
```

```python
svd = SVD()
model_selection.cross_validate(svd, us_restaurant_review_lasvegas_nv_data, verbose = True)
```

```python
nmf = NMF()
model_selection.cross_validate(nmf, us_restaurant_review_lasvegas_nv_data, verbose = True)
```

# Model Tuning and Evaluation

### SVDpp

#### 'lr_all':[.001,.01], 'reg_all':[.1,.5] cv=3

```python
parameters_grid = {'lr_all':[.001,.01], 'reg_all':[.1,.5]}
gridsearch = model_selection.GridSearchCV(SVDpp, parameters_grid, measures=['rmse','mae'], cv=3)
gridsearch.fit(us_restaurant_review_lasvegas_nv_data)
print(gridsearch.best_params['rmse']) #combination of lr_all and reg_all that produced best RMSE score
```

```python
svdpp_tuned = SVDpp(lr_all= .01, reg_all = .1)
model_selection.cross_validate(svdpp_tuned, us_restaurant_review_lasvegas_nv_data, verbose = True)
```

### When cv is increased to 5, the lr_all and reg_all with the best RMSE score is still the same.
#### 'lr_all':[.001,.01], 'reg_all':[.1,.5] cv=5

```python
parameters_grid = {'lr_all':[.001,.01], 'reg_all':[.1,.5]}
gridsearch = model_selection.GridSearchCV(SVDpp, parameters_grid, measures=['rmse','mae'], cv=5)
gridsearch.fit(us_restaurant_review_lasvegas_nv_data)
print(gridsearch.best_params['rmse']) #combination of lr_all and reg_all that produced best RMSE score
```
