```python
import pandas as pd
import numpy as np
from surprise import KNNBasic, evaluate
from surprise import Dataset, Reader
```

```python
us_restaurant_review_charlotte_nc = pd.read_csv("yelp_dataset/us_restaurant_review_charlotte_nc.csv")
```

```python
us_restaurant_review_charlotte_nc
```

```python
reader = Reader(rating_scale = (1, 5))
us_restaurant_review_charlotte_nc_data = Dataset.load_from_df(us_restaurant_review_charlotte_nc[['user_id', 'business_id', 'review_stars']], reader)
```

```python
us_restaurant_review_charlotte_nc_data
```

```python
charlotte_nc_data_train = us_restaurant_review_charlotte_nc_data.build_full_trainset()
```

```python
sim_options = {
    'name': 'cosine',
    'user_based': False
}

knn = KNNBasic(sim_options = sim_options)

knn.fit(charlotte_nc_data_train)
```

```python
charlotte_nc_data_test = charlotte_nc_data_train.build_anti_testset()
predictions = knn.test(charlotte_nc_data_test)
```

```python
predictions
```

```python
# Split us_restaurant_review_lasvegas_nv_data into 5 folds
from surprise import SVD, evaluate
from surprise import NMF

us_restaurant_review_lasvegas_nv_data.split(n_folds=5)

svd = SVD()
evaluate(svd, us_restaurant_review_lasvegas_nv_data, measures=['RMSE'])

nmf = NMF()
evaluate(nmf, us_restaurant_review_lasvegas_nv_data, measures=['RMSE'])
```
